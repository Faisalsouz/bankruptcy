```python
import pandas as pd
```

```python
# read data
data = pd.read_csv('data.csv', dtype=object)

data.head()
```

```python
# extract the healthy dataframe
healthy = data[data['Bankrupt'] == '0']

healthy.head()
```

```python
# read the bankruptcy dates
dates = pd.read_csv('bankruptcy_dates.csv', dtype=object, index_col=0)
# sort by date
dates.sort_values(by=['Date'], inplace=True)

dates
```

```python
# replace all NAN values with 0 and non-NANs with 1
healthy_nan = healthy.notnull().astype('int')
# except for the instruments and the year
healthy_nan['Instrument'] = healthy['Instrument']
healthy_nan['Year'] = healthy['Year']

print(list(healthy_nan.columns))
print()
print(len(list(healthy_nan.columns)))

healthy_nan.head()
```

```python
healthy_test = healthy_nan.groupby(['Instrument', 'Year']).sum()
healthy_test['no_NAN_values'] = healthy_test.sum(axis=1)

healthy_test
```
