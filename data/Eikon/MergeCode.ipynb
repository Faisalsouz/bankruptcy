```python
import pandas as pd
import functools
```

## Read CSV Files

```python
# read csv files and add bankrupt and year columns to the dataframes

# -------------------------------------------------------------------
# Bankrupt Companies
# -------------------------------------------------------------------
# make a list to add yearly dataframes to
bankrupt_list = []

# aggregate bankrupt companies by year
for year in range(2000, 2020):
    bankrupt_df = pd.concat([pd.read_csv('data_bankrupt/' + 'bankrupt_data{}_{}.csv'.format(i+1, year),
                                          dtype=object) for i in range(3)])
    
    # add the year value to each dataframe
    bankrupt_df['Year'] = year
    
    # add the year dataframe to the list
    bankrupt_list.append(bankrupt_df)

# concatenate all yearly dataframes
bankrupt = pd.concat(bankrupt_list)

# add the bankruptcy value to the dataframe
bankrupt['Bankrupt'] = 1


bankrupt.head()
```

```python
# read csv files and add bankrupt and year columns to the dataframes

# -------------------------------------------------------------------
# Healthy Companies
# -------------------------------------------------------------------
# make a list to add yearly dataframes to
healthy_list = []

# aggregate healthy companies by year
for year in range(2000, 2020):
    healthy_df = pd.concat([pd.read_csv('data_healthy/' + 'healthy_data{}_{}.csv'.format(i+1, year),
                                          dtype=object) for i in range(30)])
    
    # add the year value to each dataframe
    healthy_df['Year'] = year
    
    # add the year dataframe to the list
    healthy_list.append(healthy_df)

# concatenate all yearly dataframes
healthy = pd.concat(healthy_list)

# add the bankruptcy value to the dataframe
healthy['Bankrupt'] = 0


healthy.head()
```

## The Big Boss of the Concatenations

```python
data = pd.concat([bankrupt, healthy])

data.head()
```

```python
data.describe(include = 'all')
```

```python
# write the data csv file
data.to_csv('data.csv', index=False)
```
