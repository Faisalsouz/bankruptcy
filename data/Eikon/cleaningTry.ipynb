```python
datan = data.drop(columns=['NAICS Industry Group Name', 'Delisted Quote Flag', 'Company Common Name',
                   'CIK Number', 'Instrument Is Active Flag'])
```

```python
test = datan.dropna()
test['Bankrupt'] = test['Bankrupt'].apply(lambda x: int(x))

test = test.groupby('Instrument').count()
```

```python
test
```

```python
test2 = test.loc[test['Bankrupt'] == 20].reset_index()
test2
```

```python
no_nans_CIK_list = test2['Instrument'].to_list()
print(len(no_nans_CIK_list))
type(no_nans_CIK_list[0])
```

```python
print(data.shape)
type(data.Instrument[1])
```

```python
data_test = data.loc[data['Instrument'].isin(no_nans_CIK_list)]
data_test = data_test.dropna()
data_test.sort_values(by=['Instrument', 'Year'], inplace=True)
data_test = data_test.reset_index().drop(columns=['index'])

data_test_bankrupt = data_test[data_test.Bankrupt == '1']
data_test_healthy = data_test[data_test.Bankrupt == '0']
```

```python
if not os.path.exists('small_test_sets'):
    os.makedirs('small_test_sets')
```

```python
data_test_bankrupt.to_csv('small_test_sets/data_test_bankrupt.csv', index=False)

data_test_bankrupt
```
