```python
import numpy as np
import pandas as pd
import pickle
```

```python
data = pd.read_csv('datum_zeros.csv')
data.sample(10)
```

```python
# check whether the following two columns are exactly the same
all(data['Tangible Book Value - Banks'] == data['Tangible Book Value - Reported'])
```

```python
# so we drop one of the two
data.drop(columns=['Tangible Book Value - Banks'], inplace=True)

data.sample(10)
```

```python
print(any(data < 0))
print()

for col in data.columns:
    print('Negative values in', col, any(data[col] < 0))
```

```python
# make a copy of the dataframe
norm = data.copy()

# normalize data
for col in norm.columns:
    scaled = (norm[col] - norm[col].min()) / (norm[col].max() - norm[col].min())
    
    # scale value to [-1,1] for series with negative values    
    if any(norm[col] < 0):
        norm[col] = 2 * scaled - 1
    # scale values to [0,1] for positive series
    else:
        norm[col] = scaled

norm.Instrument = data.Instrument
norm.Year = data.Year
```

```python
norm.sample(10)
```

## Don't forget to add the fucking labels

```python
# read CIK list separately for bankrupt and healthy companies
# so to add labels later

with open('CIK_list_bankrupt.txt', 'rb') as fp:
    bankrupt = pickle.load(fp)
bankrupt_CIK = [int(cik) for cik in bankrupt]

with open('CIK_list_healthy.txt', 'rb') as fp:
    healthy = pickle.load(fp)
healthy_CIK = [int(cik) for cik in healthy]
```

```python
norm['Bankrupt'] = np.where(norm.Instrument.isin(bankrupt_CIK), 1, 0)

norm.sample(10)
```
