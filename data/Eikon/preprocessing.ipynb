```python
# read the full dataset for filling the NAN rows
data = pd.read_csv('data.csv')
# drop extra columns
data.drop(columns=['NAICS Industry Group Name', 'Delisted Quote Flag', 'Company Common Name',
                                    'CIK Number', 'Instrument Is Active Flag', 'Depreciation And Amortization'], inplace=True)
data.head()
```

```python
dummy = datum.notnull().astype('int')

dummy.head()
```

```python
# make a dataframe of indeces with NANs in them
# and also recording the NAN columns for each
nans = pd.DataFrame(columns=['CIK', 'Year', '#NANs', 'Columns'])

for ind in dummy.index:
    nan_no = 0
    nan_list = []
    for col in dummy.columns:
        if not dummy.loc[ind][col]:
            nan_no += 1
            nan_list.append(col)

    new_row = {'CIK': datum.loc[ind].Instrument, 'Year': datum.loc[ind].Year, '#NANs': nan_no, 'Columns': nan_list}
    nans = nans.append(new_row, ignore_index=True)
    
nans.head()
```

```python
nanz = nans[nans['#NANs'] != 0]
print(nanz.shape)
print(nanz['#NANs'].value_counts())
```

```python
nanz.head()
```

```python
nan_cik_list = list(set(nanz.CIK.to_list()))
len(nan_cik_list)
```

```python
# for rows with 1 missing value
nan_1 = nanz[nanz['#NANs'] == 1]

for ind in nan_1.index:
    cik = nan_1.loc[ind].CIK
    year = nan_1.loc[ind].Year
    factor = nan_1.loc[ind].Columns[0]
    ind_datum = list(datum[(datum.Instrument == cik) & (datum.Year == year)].index)[-1]
    df_cik = data[data.Instrument == cik].sort_values(by=['Year'])
    ind_data = list(df_cik[(df_cik.Instrument == cik) & (df_cik.Year == year)].index)[-1]
    series = df_cik[factor]
    artifactor = series.interpolate()    
    mis_val = artifactor[ind_data] 
    datum.set_value(ind_datum, factor, mis_val)
    
```
