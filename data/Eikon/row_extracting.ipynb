```python
import numpy as np
import pandas as pd

import pickle
```

```python
# read the dictionary of CIK codes
with open('CIK_year.txt', 'rb') as fp:
    CIK = pickle.load(fp)
    

# convert all CIK codes in the dictionary to strings
#for key, value in CIK.items():
#    CIK[key] = [str(cik) for cik in value]
            

# read the csv of data
data = pd.read_csv('data.csv') #, dtype={'Instrument': 'str'})
```

```python
unique_CIKs = len(list(set(data.Instrument.to_list())))
print(unique_CIKs)
len(data.Instrument.unique())
```

```python
# creat a list to add all CIK codes to
CIK_liste = []

# iterating over the dictionary items and add CIK lists to our list
for key, value in CIK.items():
    CIK_liste += value

# remove duplicated elements
CIK_list = list(set(CIK_liste))

print('There are', len(CIK_list), 'CIK codes in our list.')

CIK_list
```

```python
len(CIK_liste)
```

```python
# drop non-quantitative columns
data_cl = data.drop(columns=['NAICS Industry Group Name', 'Delisted Quote Flag', 'Company Common Name',
                                    'CIK Number', 'Instrument Is Active Flag', 'Bankrupt'])
# check on the data shape
print(data_cl.shape)

data_cl.head()
```

```python
# check data format for CIK codes in df and dict
print(type(data_cl.Instrument[1]))
type(CIK_list[1])
```

```python
# drop the all companies but the ones in the CIK list
data_cl = data_cl[data_cl.Instrument.isin(CIK_list)]

print('Data df shape:', data_cl.shape, '\n')

print('If every thing is okay, then the number of unique CIK in semi-clean data,', len(data_cl.Instrument.unique()),
      '\nshould be equal to the number of CIK codes in the dictionary:', len(CIK_list))

print()
print('Is that true?')
print(str(len(data_cl.Instrument.unique()) == len(CIK_list)) + '!')


data_cl.head()
```

```python
# iterate over the dictionary, and for each year concatanate 
# the corresponding five years for each company in the list
# to the final dataframe: datum.

# create an issue log
issues = ''

# create a dataframe to record the issue
df_issues = pd.DataFrame(columns=['CIK', 'Year', '#rows'])

# create an empty df with the data columns
datum = pd.DataFrame(columns=data_cl.columns)

# iterate over the dictionary
for key, value in CIK.items():
    # for each company (cik) in the list of the year (key)
    for cik in value:
        # take rows of the data for that cik code and the same year as key
        df = data_cl[(data_cl.Instrument == cik) & (data_cl.Year == key)]
        if df.shape[0] > 1:
            issues += '\n ' + str(cik) + ' for the year ' + str(key) + ' has ' + str(df.shape[0]) + ' rows.'
            new_row = {'CIK': cik, 'Year': key, '#rows': df.shape[0]}
            df_issues = df_issues.append(new_row, ignore_index=True)
        datum = pd.concat([datum, df])

print(datum.shape)
datum.head()
```

```python
# drop the duplicate rows
datum.drop_duplicates(inplace=True)

print(datum.shape)

datum.sort_values(by=['Instrument', 'Year'])
```

```python
datum['dummy'] = 1
dummy = datum[['Instrument', 'Year', 'dummy']]

dummy.head()
```

```python
dummy = dummy.groupby(['Instrument', 'Year']).sum().reset_index()
dummy
```

```python
dummy_list = dummy.dummy.to_list()
for i in range(10):
    print('Do we have', i, 'repititive rows in the list?', (i in dummy_list))
```

```python
# how many double rows are still in the datafram?
dummy_list.count(2)
```

```python
dup_rows = dummy[dummy.dummy == 2]
dup_rows.head()
```

```python
datum.drop(columns=['dummy'], inplace=True)
datum.head()
```

```python
for ind in dup_rows.index:
    temp = datum[(datum.Instrument == dup_rows.loc[ind].Instrument) & (datum.Year == dup_rows.loc[ind].Year)]
    temp = temp.notnull().astype('int')
    temp['noNANs'] = temp.sum(axis=1)
    inds = list(temp.index)
    if temp.loc[inds[0]].noNANs > temp.loc[inds[1]].noNANs:
        datum.drop(index=inds[1], inplace=True)
    else:
        datum.drop(index=inds[0], inplace=True)

print(datum.shape)
datum.head()
```

```python
datum.sort_values(by=['Instrument', 'Year'], inplace=True)
datum
```

```python
# the dataframe with no extra double row
datum.to_csv('datum.csv', index=False)
```
