```python
with open('urgent_h.txt', 'rb') as f:
    healthy_cik = pickle.load(f)

with open('urgent_b.txt', 'rb') as f:
    bankrupt_cik = pickle.load(f)

data = pd.read_csv('/Users/user/Documents/Bankruptcy/bankruptcy/data/processed/data.csv')
data.head()
```

```python
sample = 0
dataset_healthy = {}
problems_healthy = []

for key in healthy_cik:
    start = int(key[6:10])
    end =  int(key[11:15])
    years = list(range(start, end+1))
    ciks = healthy_cik[key]
    for cik in ciks:
        df = pd.DataFrame(columns=data.columns)
        for year in years:
            newrow = data[(data.CIK == cik) & (data.Year == year)]
            df = df.append(newrow, ignore_index=True)
        
        matrix = df.drop(columns=['CIK', 'Year','Bankrupt'])
            
        if matrix.shape[0] == 5:
            sample_dic = {}
            sample_dic['CIK'] = cik
            sample_dic['Period'] = key
            sample_dic['data'] = matrix.as_matrix()
            sample_dic['label'] = 0
            dataset_healthy[sample] = sample_dic
        else:
            problems_healthy.append([cik, df.Year.to_list(), key])
        sample += 1


with open('dataset_healthy.txt', 'wb') as f:
    pickle.dump(dataset_healthy, f)

with open('problems_healthy.txt', 'wb') as f:
    pickle.dump(problems_healthy, f)
```

```python
sample = 0
dataset_bankrupt = {}
problems_bankrupt = []

for key in bankrupt_cik:
    start = int(key[6:10])
    end =  int(key[11:15])
    years = list(range(start, end+1))
    ciks = bankrupt_cik[key]
    for cik in ciks:
        df = pd.DataFrame(columns=data.columns)
        for year in years:
            newrow = data[(data.CIK == cik) & (data.Year == year)]
            df = df.append(newrow, ignore_index=True)
        
        matrix = df.drop(columns=['CIK', 'Year','Bankrupt'])
            
        if matrix.shape[0] == 5:
            sample_dic = {}
            sample_dic['CIK'] = cik
            sample_dic['Period'] = key
            sample_dic['data'] = matrix.as_matrix()
            sample_dic['label'] = 1
            dataset_bankrupt[sample] = sample_dic
        else:
            problems_bankrupt.append([cik, df.Year.to_list(), key])
        sample += 1

        
with open('dataset_bankrupt.txt', 'wb') as f:
    pickle.dump(dataset_bankrupt, f)

with open('problems_bankrupt.txt', 'wb') as f:
    pickle.dump(problems_bankrupt, f)
```
