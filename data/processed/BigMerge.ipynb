```python
import pandas as pd
```

## Read Word-Distance CSV

```python
word_dist = pd.read_csv('word_distances_200.csv')

word_dist.head()
```

```python
word_dist.fillna(0, inplace=True)
word_dist.head()
```

```python
word_dist.shape
```

## Read Financial Factors CSV

```python
fin_fact = pd.read_csv('fin_facts_fill_norm.csv')
fin_fact.head()
```

```python
fin_fact.shape
```

```python
fin_fact.rename(columns={'Instrument': 'CIK'}, inplace=True)
fin_fact.head()
```

## Check Two Dataset CIK Lists

```python
print('The unique number of companies in the financial factors dataset:')
print(len(set(fin_fact.CIK.to_list())))

print('\nThe unique number of companies in the word-distance dataset:')
print(len(set(word_dist.CIK.to_list())))
```

## Merging

```python
fin_fact.set_index(['CIK', 'Year'], inplace=True)
word_dist.set_index(['CIK', 'Year'], inplace=True)

data = pd.concat([fin_fact, word_dist], axis=1)
```

```python
data.head()
```

```python
data.fillna(0, inplace=True)
data.head(20)
```

```python
data.reset_index(inplace=True)
data
```

## Write the CSV File

```python
data.to_csv('data.csv', index=False)
```
